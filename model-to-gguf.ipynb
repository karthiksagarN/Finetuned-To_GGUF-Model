**5.Pushing the model a Repository on Hugging Face Hub and Upload the GGUF Model**

```python
from huggingface_hub import HfApi
import os

# Set the token as an environment variable (recommended for security)
os.environ["HUGGING_FACE_HUB_TOKEN"] = "PUT YOUR Huggging FACE Token here"

api = HfApi()
#change your-own-account/Indic-gemma-2b-finetuned-sft-Navarasa-2.0.gguf

model_id = "karthiksagarn/model_filename.gguf"
api.create_repo(model_id, exist_ok=True, repo_type="model") # Use the api object to create the repo
api.upload_file(
    path_or_fileobj="model_filename.gguf",
    path_in_repo="model_filename.gguf",
    repo_id=model_id,
)
```
